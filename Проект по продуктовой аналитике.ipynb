```python
first_install = df[df.event=='app_install'].groupby('date', as_index=False).device_id.nunique()
```

```python
first_install
```

```python
first_purchase = df[df.event=='purchase'].groupby('device_id', as_index=False).date.min()
```

```python
first_purchase
```

```python
first_app_install = df[df.event=='app_install'].groupby('device_id', as_index=False).date.min()
```

```python
first_app_install
```

```python
al = first_app_install.merge(first_purchase, how='inner', on='device_id')
```

```python
al.rename(columns={'date_x':'date_app_install', 'date_y':'date_purchase'}, inplace=True)
```

```python
al
```

```python
al['delta'] = al['date_purchase'] - al['date_app_install']
```

```python
al
```

```python
al = al.loc[al['delta']<pd.Timedelta(8,'D')]
```

```python
al
```

```python
alconv = al.groupby('date_app_install', as_index=False).device_id.nunique().sort_values('device_id')
```

```python
alconv
```

```python
alconv.rename(columns = {'date_app_install':'date'}, inplace=True)
```

```python
alconv
```

```python
all = first_install.merge(alconv, how='inner', on='date')
```

```python
all
```

```python
all['conv'] = all['device_id_y']/all['device_id_x']*100
```
